```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.optimize import curve_fit
from scipy import integrate
import math
```

```python
x4a, y4a = np.loadtxt('/Users/liuhaoyun/Desktop/Uni/Year_2/Year_2_Lab/Year_2_lab_waves/PART2_Thermal_Waves/data_sets/thermal_4min_a.txt', unpack = True, skiprows = 3)
x4b, y4b = np.loadtxt('/Users/liuhaoyun/Desktop/Uni/Year_2/Year_2_Lab/Year_2_lab_waves/PART2_Thermal_Waves/data_sets/thermal_4min_b.txt', unpack = True, skiprows = 3)

def square(t):
    A = 50
    tau = 240
    return A*signal.square(2 * np.pi * t/tau)+50 

def T_n1(t):
    T_1 = 50 + (200 / np.pi) * np.sin(np.pi * t /120)
    return T_1

def sine(t, A, f, phi, C):
    return A * np.sin(2 * np.pi * f * t/10 + phi) + C

p0 = [10, 1/240, np.pi, 30]
popt, pcov = curve_fit(sine, x4a, y4a, p0)

print("Amplitude: %.3f" %popt[0],"+/- %.3f" %np.sqrt(pcov[0,0]))
print("Frequency: %.3f" %popt[1],"+/- %.3f" %np.sqrt(pcov[1,1]))
print("Phase shift: %.3f" %popt[2],"+/- %.3f" %np.sqrt(pcov[2,2]))

def D_TF(frequency_n, delta_r, gamma_n):
    return (2*np.pi*frequency_n*delta_r**2)/(2*np.log(gamma_n)**2)

def D_PL(frequency_n, delta_r, phi):
    return (2*np.pi*frequency_n*delta_r**2)/(2*phi**2)

frequency_n = popt[1]
delta_r = 7.82
gamma_n = popt[0]/(200/np.pi)
phi = popt[2]

print("D_TF: %.3f" %D_TF(frequency_n, delta_r, gamma_n),'mm^2 s^-1')
print("D_PL: %.3f" %D_PL(frequency_n, delta_r, phi),'mm^2 s^-1')

plt.figure(figsize=(10,6))
plt.plot(x4a, sine(x4a, *popt), label='Fitted for data a',linewidth=7.0)
plt.plot(x4a,y4a,label='Dataset A')
plt.plot(x4b,y4b,label='Dataset B')
plt.plot(x4a,T_n1(x4a/10),label='Fundamental of square')
plt.plot(x4a,square(x4a/10),label='Ideal Sqaure')
plt.title("Temperature 4 min")
plt.xlabel("Time ($10^{-1}s$)")
plt.ylabel("Temperature ($ \degree C$)")
plt.xlim(0,9600)
plt.legend()
plt.grid()
plt.show()
```

```python
# Fourier analysis
t = x4a
data = y4a
period = 2400 #(in ms plz)
number_of_terms = 5 #(n)

# Fourier coefficients
def an(t, data, n):
    an_coefficients = integrate.cumtrapz(data * np.cos(2*np.pi*n*t/period)*2/(period*4),t,initial = 0)
    return an_coefficients[-1]

def bn(t, data, n):
    bn_coefficients = integrate.cumtrapz(data * np.sin(2*np.pi*n*t/period)*2/(period*4),t,initial = 0)
    return bn_coefficients[-1]

def Cosine(n):
    return np.cos(2 * np.pi * n * t/period)

def Sine(n):
    return np.sin(2 * np.pi * n * t/period)

sum = 0
for n in range(1,number_of_terms+1):
    sum += an(t,data,n)*Cosine(n) + bn(t,data,n)*Sine(n)
    print("n =",n)
    print("a:",an(t, data ,n))
    print("b:",bn(t, data ,n))
    print('beta:',np.sqrt(an(t, data ,n)**2+bn(t, data ,n)**2))
    print('phase lag',-math.atan2(an(t, data ,n),bn(t, data ,n)))

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(t, data, color='yellow',label="Original Signal")
plt.plot(t,an(t,data,0)/2+sum,color='red',label="Fourier Series Approximation", linestyle='dashed')
plt.title("Fourier Series Approximation")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```
