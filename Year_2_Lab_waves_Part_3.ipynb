```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
# Read data
df = pd.read_csv("/Users/liuhaoyun/Desktop/Uni/Year_2/Year_2_Lab/Year_2_lab_waves/Year_2_Lab_part_3_task_3.6.2.csv")
df
```

```python
k = df.iloc[:,10]
omega = df.iloc[:,11]

v_phase = omega/k
#print(v_phase)

m, b = np.polyfit(k , omega, 1)
print(m)
print(b)

plt.plot(k,m*k+b,label='fitted line')
plt.plot(k, omega,'.')
plt.title('$\omega$ vs k')
plt.xlabel("k")
plt.ylabel("$\omega$")
plt.grid()
plt.legend()
plt.show()
```

```python
frequency = df.iloc[:,2]
deltak = df.iloc[:,12]
deltaomega = df.iloc[:,13]
v_group = deltaomega/deltak
plt.plot(frequency,v_phase,"x",label = "v_phase")
plt.plot(frequency,v_group,'.',label = 'v_group')
plt.grid()
plt.legend()
plt.show()
```
